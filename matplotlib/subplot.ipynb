display the multiple plots-with subplot() you can draw multiple plots in one figure

```python
import matplotlib.pyplot as plt
import numpy as np
```

column wise display of two graphs

```python
#plot 1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
#here 1 2 1 represent- 1:row, 2:column ,1:this plot is the 1 plot
plt.subplot(1,2,1)
plt.plot(x,y)

#plot 2
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()

#1 row 2 column
```

row wise display of two graphs

```python
#plot 1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
#here 1 2 1 represent- 1:row, 2:column ,1:this plot is the 1 plot
plt.subplot(2,1,1)
plt.plot(x,y)

#plot 2
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,1,2)
plt.plot(x,y)

plt.show()
```

now we will draw a challenge of 6 plots

```python
#plot 1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
#here 1 2 1 represent- 1:row, 2:column ,1:this plot is the 1 plot
plt.subplot(2,3,1)
plt.plot(x,y)
plt.title("graph 1")


#plot 2
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,3,2)
plt.plot(x,y)
plt.title("graph 2")

#plot 3
x=np.array([0,1,2,3])
y=np.array([10,2,30,4])
plt.subplot(2,3,3)
plt.plot(x,y)
plt.title("graph 3")

#plot 4
x=np.array([0,1,2,3])
y=np.array([1,20,3,40])
plt.subplot(2,3,4)
plt.plot(x,y)
plt.title("graph 4")

#plot 5
x=np.array([0,1,12,3])
y=np.array([10,20,30,4])
plt.subplot(2,3,5)
plt.plot(x,y)
plt.title("graph 5")

#plot 6
x=np.array([0,1,2,13])
y=np.array([10,2,30,40])
plt.subplot(2,3,6)
plt.plot(x,y)
plt.title("graph 6")

plt.show()
```
