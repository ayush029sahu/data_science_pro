```python
from tkinter import Y
import matplotlib.pyplot as plt
import numpy as np
```

pyplot is the utility working and styling is stored in pyplot
plot() function is used to draw points or markers in a diagram

```python
xaxis=np.array([0,6])
yaxis=np.array([0,250])
plt.plot(xaxis,yaxis)
plt.show()
```

now plotting without line

```python
xaxis=np.array([1,8])
yaxis=np.array([3,10])
plt.plot(xaxis,yaxis,'o')
#here 'o' is the marker used to replace line with points
plt.show()
```

zig zag graph

```python
x=np.array([1,2,6,8])
y=np.array([3,8,1,10])
plt.plot(x,y)
plt.show()
```

if the  values do not provide by the user then python plot() function by default asssign values to the axis in comparison with other axis through its interval

lets take default x-points

```python
y=np.array([3,8,1,10,5,7])
plt.plot(y)
plt.show()
```

markers-you can use argument marker to emphasis each point with specified marker

```python
y=np.array([3,8,1,10,5,7])
plt.plot(y,marker='*')
plt.show()
```

format strings "fmt"- marker|line|color

```python
y=np.array([3,8,1,10])
plt.plot(y,'o:r')
#'o-symbol :-dotted line r-red color'
#':-dotted line , - -> solid line ,-- ->dashed line'
plt.show()
```

marker size-ms

```python
y=np.array([3,8,1,10])
plt.plot(y,'o-b',ms='20')
plt.show()
```

marker edge color- mec

```python
y=np.array([3,8,1,10])
plt.plot(y,'o-',ms='10',mec='r')
plt.show()
```

mfc-marker face color

```python
y=np.array([3,8,1,10])
plt.plot(y,marker='o',ms='10',mec='y',mfc='r')
plt.show()
```

lines

line style or ls-used to change the style of the plotted line

```python
y=np.array([3,8,1,10])
plt.plot(y,marker='o',linestyle='dotted',ms='20',mec='y',mfc='g')
plt.show()
```

line color-c

```python
y=np.array([3,8,1,10])
plt.plot(y,color='r')
plt.show()
```

line width or lw

```python
y=np.array([3,8,1,10])
plt.plot(y,ls='-',c='b',lw='5')
plt.show()
```

example for multiple line

```python
x=np.array([1,2,5,3])
y=np.array([3,8,1,10])
plt.plot(x)
plt.plot(y)
plt.show()
```

labels and titles

create labels for a plot
and title

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.xlabel("Average oxygen")
plt.ylabel("Our calorie")
plt.plot(x,y,marker='o',ls='-',mfc='y',c='g')
plt.title("Health Monitor")
plt.show()
```

now we will set the font family and property for title and labels by using 'fontdict'

```python
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.xlabel("Average oxygen",fontdict=font2)
plt.ylabel("Our calorie",fontdict=font2)
plt.plot(x,y,marker='o',ls='-',mfc='y',c='g')
plt.title("Health Monitor",fontdict=font1)


plt.show()
```

changing the location of the labels

```python
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.xlabel("Average oxygen",fontdict=font2,loc='right')
plt.ylabel("Our calorie",fontdict=font2)
plt.plot(x,y,marker='o',ls='-',mfc='y',c='g')
plt.title("Health Monitor",fontdict=font1,loc='center')


plt.show()
```

grid lines-used to give proper representation of the graph using grid lines to a plot via grid() function

```python
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.xlabel("Average oxygen",fontdict=font2,loc='right')
plt.ylabel("Our calorie",fontdict=font2)
plt.title("Health Monitor",fontdict=font1)
plt.plot(x,y,marker='o',ls='-',mfc='y',c='g')
plt.grid()



plt.show()
```

now we will specify which grid lines to display via a x axis or y axis. (legal values are x and y and for default it is both)

```python
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.xlabel("Average oxygen",fontdict=font2,loc='right')
plt.ylabel("Our calorie",fontdict=font2)
plt.title("Health Monitor",fontdict=font1)
plt.plot(x,y,marker='o',ls='-',mfc='y',c='g')
plt.grid(axis='x')



plt.show()
```

changing the grid line color using line property as the grid is line 

```python
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.xlabel("Average oxygen",fontdict=font2,loc='right')
plt.ylabel("Our calorie",fontdict=font2)
plt.title("Health Monitor",fontdict=font1)
plt.plot(x,y,marker='o',ls='-',mfc='y',c='g')
plt.grid(color='green',ls='--',lw=0.5)



plt.show()
```
