creating bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.show(),
```

now we will create a horizontal bar

```python
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.barh(x,y)
plt.show(),
```

colouring of the bar,width of bar
\but for horizonatal bar we use height rather width

```python
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.bar(x,y,color='red',width=0.1)
plt.show(),
```

PIE CHART by pie function

```python
y=np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

labeling the sections of the pie chart
pie chart do anticlock wise motion 
and it has its starting point at x-axis but we can change it

```python
y=np.array([35,25,25,15])
x=np.array(['apple','mango','banana','grapes'])
plt.pie(y,labels=x)
plt.show()
```

changing the starting point of the bar chart
using the startangleparameter

```python
y=np.array([35,25,25,15])
x=np.array(['apple','mango','banana','grapes'])
plt.pie(y,labels=x,startangle=90)
plt.show()
```

how to explode the pie chart by a value.

```python
y=np.array([35,25,25,15])
x=np.array(['apple','mango','banana','grapes'])
ex=([0.2,0,0,0])
#here 0.2 is the distance from center

plt.pie(y,labels=x,explode=ex)
plt.show()
```

the shadow parameter for creativity

```python
y=np.array([35,25,25,15])
x=np.array(['apple','mango','banana','grapes'])
ex=([0.2,0,0,0])
#here 0.2 is the distance from center

plt.pie(y,labels=x,explode=ex,shadow=True)
plt.show()
```

colouring of each wedges

```python
y=np.array([35,25,25,15])
x=np.array(['apple','mango','banana','grapes'])
clr=(["green","hotpink","blue","black"])

plt.pie(y,labels=x,colors=clr)
plt.show()
```

we can also add the legend- list of explanation

```python
y=np.array([35,25,25,15])
x=np.array(['apple','mango','banana','grapes'])
clr=(["green","hotpink","blue","black"])

plt.pie(y,labels=x,colors=clr)
plt.legend()
plt.show()
```
