scatter-the scatter () func plots one dot for each observation

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)
plt.show()
```

now we will compare 2 plots on same figure

```python
#day1 the age and speed
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)

#day2 the age and speed
x=np.array([5,7,8,7,2,1,12,9,14,11,12,19,6])
y=np.array([99,86,87,8,11,6,13,8,94,7,7,5,86])
plt.scatter(x,y)



plt.show()
```

markers

```python
#day1 the age and speed
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y,color='hotpink')

#day2 the age and speed
x=np.array([5,7,8,7,2,1,12,9,14,11,12,19,6])
y=np.array([99,86,87,8,11,6,13,8,94,7,7,5,86])
plt.scatter(x,y,color='green')



plt.show()
```

we can change color of each dot

```python
x=np.array([5,7,8,3,5,6,10])
y=np.array([99,86,87,24,22,15,34])
#create a variable with different color for each dot
colour=(['red','green','hotpink','yellow','grey','purple','black'])
plt.scatter(x,y,color=colour)
plt.show()
```

now we will create a color array and specify a colormap in scatter plot

```python
x=np.array([5,7,8,3,5,6,10])
y=np.array([99,86,87,24,22,15,34])
#create a variable with different color for each dot
colour=([0,10,20,24,40,25,32])
plt.scatter(x,y,c=colour,cmap='viridis')
plt.show()
```

you can also include color bar 

```python
x=np.array([5,7,8,3,5,6,10])
y=np.array([99,86,87,24,22,15,34])
#create a variable with different color for each dot
colour=([0,10,20,24,40,25,32])
plt.scatter(x,y,c=colour,cmap='viridis')
plt.colorbar()
plt.show()
```

we can also change the size

```python
x=np.array([5,7,8,3,5,6,10])
y=np.array([99,86,87,24,22,15,34])
#create a variable with different color for each dot
sizes=np.array([20,50,10,20,30,60,70])
plt.scatter(x,y,s=sizes)
plt.show()
```

alpha-you can adjust the transpracy of the dots

```python
x=np.array([5,7,8,3,5,6,10])
y=np.array([99,86,87,24,22,15,34])
#create a variable with different color for each dot
sizes=np.array([120,150,110,120,130,160,70])
plt.scatter(x,y,s=sizes,alpha=0.5)
plt.show()
```

```python
x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
sizes=10*np.random.randint(100,size=(100))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='nipy_spectral')
plt.colorbar()
plt.show()
```
