using digit data set of hand written images to conclude the digit

```python
import pandas as pd
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.pink()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
digits.data[:5]
```

```python
df=pd.DataFrame(digits.data)
df.head()
```

```python
df['target']=digits.target
df.head()
```

the above df shows that the first row in df is mapped for the target 0 same as with other rows and their target 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(['target'],axis='columns'),digits.target,test_size=0.2)
```

```python
len(x_test)
```

ensemble is a term use when we are dealing with multiple algorithms to predict the outcome in random forest to create multliple decision tree

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=200)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
