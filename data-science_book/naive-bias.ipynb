conditional probability-p[a/b]=probability of event A knowing that event B has already occurred
p(a/b)=p(b/a)*p(a)/p(b)

q1=pick a random card while you know that its a diamond then.now what is the probability of that card being a queen?

ans--using the bayes theorem of conditional probability
p(queen/diamond)=p(diamond/queen)*p(queen)/p(diamond)

```python
import pandas as pd
df=pd.read_csv("titanic.csv")
df.head()
```

we need to drop unnecessary data from the data frame

```python
df.drop(['PassengerId','Name','Parch','Ticket','Cabin','Embarked','SibSp'],axis='columns',inplace=True)
df.head()
```

```python
target=df.Survived
inputs=df.drop('Survived',axis='columns')
```

since on the sex column it contains string so we use need to convert it into value

```python
dummies=pd.get_dummies(inputs.Sex)
dummies=dummies.astype(int)
dummies.head()
```

```python
inputs=pd.concat([inputs,dummies],axis='columns')
inputs.head()
```

```python
inputs.drop('Sex',axis='columns',inplace=True)
inputs.head()
```

to find that our complete data frame does contains any null value in any columns 

```python
inputs.columns[inputs.isna().any()]
```

above it return age column contains a null value

check which rows it contains null value and then handle it

```python
inputs.Age[:10]
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

creating naive bayes 

inside naive bayes there are many models presents amongst whcih we will use the gaussianNB model

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
