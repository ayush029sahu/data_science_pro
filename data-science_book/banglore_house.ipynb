if we pass the value of total_sqft inside the is_float function if it is a float value it will return true if it is in range it will return false

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

~ this is a negate function which will return the false value 
which are ranges

```python
df3[~df3['total_sqft'].apply(is_float)].head(10)
```

in above dataframe we can see that some values are in sq.meter,perch etch we need to also preprocess it 

now we will write a python function which will replace the average value with range values which it takes as a string input

```python
def avrg(x):
    tokens=x.split('-')
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

If the length of the list tokens is 2, this means the string x contained exactly one hyphen, resulting in two parts

```python
df4=df3.copy()
df4['total_sqft']=df4['total_sqft'].apply(avrg)
df4.head(20)
```

wehen divide the (total_sqft/bhk) and calculate its mean and then check its deviated value we would able to find which all datapoints fail fit model since they got error

```python
df6=df5[~(df5.total_sqft/df5.bhk<300)]
df6.shape
```

now working on the price column we do have to remove the datapoints which are extremes like extrememly low or extremely high

```python
df6.price_per_sqft.describe()
```

since 267.82 is very low and 176470 is high to make generalized model we have to remove these extreme cases

```python
def remove_pps_outlier(df):
    df_out=pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m=np.mean(subdf.price_per_sqft)
        st=np.std(subdf.price_per_sqft)
        reduced_df=subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out=pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out    

df7=remove_pps_outlier(df6)
df7.shape    
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

here the use of the shufflesplit is that will continuesly shuffle the data set for the 5 folds every time so it do not target the same data sets all the time

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
cross_val_score(LinearRegression(),x,y,cv=cv)
```

using gridsearch technique we can find which is the best model and their parametes for our dataset

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
```

```python
def find_best_model(x,y):
    algos={
        'linear_regression':{
            'model':LinearRegression(),
            'params':{
                'fit_intercept':[True,False]
            }

        },
        'lasso':{
            'model':Lasso(),
            'params':{
                'alpha':[1,2],
                'selection':['random','cyclic']
            }

        },
        'decison_tree':{
            'model':DecisionTreeRegressor(),
            'params':{
                'criterion':['mse','friedman_mse'],
                'splitter':['best','random']
            }
        }
    }
    scores=[]
    cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
    for algo_name,config in algos.items():
        gs=GridSearchCV(config['model'],config['params'],cv=cv,return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model':algo_name,
            'best_score':gs.best_score_,
            'best_params':gs.best_params_
        })
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
find_best_model(x,y)
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(x.columns==location)[0][0]

    X = np.zeros(len(x.columns))
    X[0] = sqft
    X[1] = bath
    X[2] = bhk
    if loc_index >= 0:
        X[loc_index] = 1

    return lr.predict([X])[0]
```

```python
predict_price('1st Phase JP Nagar',1000, 2, 2)
predict_price('Electronic City Phase II',1000,2,2)
```
