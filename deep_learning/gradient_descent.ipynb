Now we start implementing gradient descent in plain python. Again the goal is to come up with same w1, w2 and bias that keras model calculated. We want to show how keras/tensorflow would have computed these values internally using gradient descent

First write couple of helper routines such as sigmoid and log_loss

```python
def sigmoid_numpy(x):
    return 1/(1+np.exp(-x))

sigmoid_numpy(np.array([12,0,1]))
```

to reduce our losses we use the log_loss

```python
def log_loss(y_true,y_predicted):
    epsilon=1e-15
    y_predicted_new=[max(i,epsilon) for i in y_predicted]
    y_predicted_new=[min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new=np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

All right now comes the time to implement our final gradient descent function !! yay !!!

here we set the values of w1 and w2 as 1 and bias=0 and the learning_rate as 0.5

```python
def gradient_descent(age,affordibility,y_true,epochs,loss_threshold):
    w1=w2=1
    bias=0
    rate=0.5
    n=len(age)
    for i in range(epochs):
        weighted_sum=w1*age+w2*affordibility+bias
        y_predicted=sigmoid_numpy(weighted_sum)
        loss=log_loss(y_true,y_predicted)
#for calculating the new w1 and w2 we first need to find the partial derivative of the slope of the function so as to adjust the global minimum
#This is the gradient of the loss function with respect to the parameter w1
#It measures the rate of change of the loss function with respect to w1
#The gradient points in the direction of the steepest increase in the loss function, so subtracting it moves the parameter in the direction that decreases the loss.
        w1d=(1/n)*np.dot(np.transpose(age),(y_predicted-y_true))
        w2d=(1/n)*np.dot(np.transpose(affordibility),(y_predicted-y_true))

#now to update the w1 and w2 after the loss calcualtion and to readjust weights to reduce loss
#using the formula w1=w1-learning rate * d/dw1
        #the updated values are after the loss calculattion
        bias_d=np.mean(y_predicted-y_true)
        w1=w1-rate*w1d
        w2=w2-rate*w1d
        bias=bias-rate*bias_d

        print(f'epochs:{i},w1:{w1},w2:{w2},bias:{bias},loss:{loss}')
#the threshold value will tell us to when to stop after finding correct parametes
        if loss<=loss_threshold:
            print(f'Loss threshold reached. Stopping training.')
            break

    return w1,w2,bias
```
