normal(gaussian) distribution -very important
random.normal() mehtod -loc(mean parameter/peak level),scale(flatness of graph/standard deviation),size

```python
from numpy import random
```

```python
x=random.normal(size=(2,3))
print(x)
#here we have generated a random normal 
#distribution of size 2*3
```

we will generate a random normal dist of size 2*3 with mean at 1 and standard deviation of 2:

```python
x=random.normal(loc=1,scale=2,size=(2,3))
print(x)
```

visualilzation of normal distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.normal(size=100),hist=True,kde=True)
plt.show()
```

the curve of a normal dist is also called as bell curve

Binomial distribution - discrete binary output 
it has 3 parameterrs - 1:n(number of trials),
2:-p(probability)
3:-size(shape-returned array)




given 10 trials for a coin whcich will generate 10 data points:

```python
x=random.binomial(n=10,p=0.5,size=10)
print(x)
```

visualization of binomial

```python
sns.distplot(random.binomial(n=5,p=0.6,size=1100),hist=True,kde=True)
plt.show()
```

difference between normal and binomial -normal (continues) and binomial(discrete)
i use 500 data points for binomial and under 100 data points for normal dist

```python
sns.distplot(random.normal(loc=50,scale=5,size=1000),hist=False,label='normal')
sns.distplot(random.binomial(n=100,p=0.5,size=1000),hist=False,label='binomial')
plt.show()
```

poission distribution - it is descrete, it assumes the pridict the occurence of the happening of the event according to its occurence and results 
it estimate how many time an event can happen
parameter-lam(number of occurence or rate),size
generate-a random 1*10 dist for the occurence of 2


```python
x=random.poisson(lam=2,size=10)
print(x)
```

visiualization of the poission dist

```python
sns.distplot(random.poisson(lam=2,size=1000),kde=False)
plt.show()
```

presenting both the pllots in same figure normal and poission

```python
sns.distplot(random.normal(loc=50,scale=7,size=100),hist=False)
sns.distplot(random.poisson(lam=50,size=100),kde=True,hist=False)
plt.show()
```

multinomial distribution-it is the generalization of the binomial dist
parameter are -n(number of posibility or outcomes),pvals(list of posibilities of the outcomes),size(shape of returned array)

Q1: draw out a sample for dice roll

```python
x=random.multinomial(n=6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6])
print(x)
#multinomial samples will not produce a single values ,they will 
#produce one value for each pvals.
```
